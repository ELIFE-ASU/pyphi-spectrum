## Notebook to find the numerical upper bound on $\Phi^{max}$ for a two-bit elementary logic gate system

```python
import pyphi
from pyphi import phi_spectrum
import numpy as np
import itertools

# n_bits = 2
# n_states = 2**n_bits

states = ['1','2','3','4']
label_state_map = {'1':[0,0],'2':[1,0],'3':[0,1],'4':[1,1]}
possible_TPM = itertools.product(states, repeat=4)

phi_values = []
for each in possible_TPM:
    print("\nANALYZING TPM...",)
    
    ## Get TPM
    TPM = []
    for key in each:
        TPM.append(label_state_map[key])
    print(TPM)

    ## Create Network object
    network = pyphi.Network(TPM, node_labels=['A','B'])

    # For each initial state, calculate the Phi values
    reachable_states = np.unique(each)
    for state in reachable_states:
        
        initial_state = label_state_map[state]
        print("\nInitial State = ",initial_state)

        nodes = ['A','B']
        subsystem = pyphi.Subsystem(network, initial_state, nodes)

        Phi_Spectrum = phi_spectrum.get_phi_spectrum(subsystem)
        All_Phi = phi_spectrum.get_Phi_MIP(Phi_Spectrum)
        print("PHI = ",All_Phi)
        phi_values.extend(All_Phi)
        
print("DONE")
```

```python
print("MINIMUM PHI VALUE = ",np.min(phi_values))
print("MAXIMUM PHI VALUE = ",np.max(phi_values))
```
